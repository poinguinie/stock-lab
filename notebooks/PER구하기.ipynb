```python
import pandas as pd
corp_info = pd.read_csv('corp.csv', header=0, encoding='cp949')
# corp_info = corp_info[['종목코드','종목명','현재가','상장주식수']]
corp_info = corp_info[['종목코드','종목명','종가','상장주식수']]

items = []
for item in corp_info['종목코드']:
    item  = format(item, '06')
    items.append(item)
    
corp_info['종목코드'] = items
corp_info['종목코드'] = 'A' + corp_info['종목코드'].apply(str)

corp_info
```

```python
corp_result = pd.read_excel('SAMPLE_2018.xlsx')
corp_result
```

```python
corp_data = corp_result[['종목코드','당기순이익']]
corp_data
```

```python
corp_total = pd.merge(corp_info, corp_data, how='left', on='종목코드')
corp_total
```

```python
corp_total["EPS"] = corp_total['당기순이익']/corp_total['상장주식수']
corp_total
```

```python
corp_total = corp_total[ pd.notnull(corp_total['EPS']) ]
corp_total = corp_total[ pd.notnull(corp_total['종가']) ]
corp_total
```

```python
corp_total["PER"] = corp_total["종가"]/corp_total["EPS"]
corp_total
```

```python
corp_filter = corp_total[(corp_total["PER"] >= 12) & (corp_total["PER"] <= 15)]
corp_filter
```
